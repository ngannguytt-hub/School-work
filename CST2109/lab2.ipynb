```python
fahrenheit = float(input("Enter the Fahrenheit temperature: "))
celsius = (fahrenheit - 32) / 1.8

print(f"Fahrenheit: {fahrenheit}°F")
print(f"Celsius: {celsius:.2f}°C")
print("Student Number: ", student_number)
```

This code asks the user to enter a temperature in Fahrenheit, then converts it to Celsius using a formula, and finally prints both temperatures and student number.

```python
student_info = {
    "Student Number" : "041154207",
    "Name" : "Ngan Nguyen",
    "Program" : "Business Intelligence System Infrastructure",
    "Courses" : ["Business Intelligence", "Machine Learning 1"],
    "Instructors" : ["Rani Adhaduk", "Umer Altaf"]
}

print("Courses I am taking: ")
for course in student_info["Courses"]:
    print (f" - {course}")

print(f"Student Name: {student_info.get("Name")}")

print("All Student information values: ")
for value in student_info.values():
    print (f" - {value}")
```

This code stores student information in a dictionary, including Student Number, Name, Program, Courses, and Instructors.
It then prints the list of courses, displays the student’s name, and finally loops through the dictionary to print all stored values.

```python
for number in range (1,11):
    if number % 2 == 0 and number % 5 == 0 : print("FizzBuzz", end=" ")
    elif number % 2 == 0 : print("Fizz", end=" ")
    elif number % 5 == 0 : print("Buzz",  end=" ")
    else : print(number,  end=" ")
```

This code loops through numbers from 1 to 10.
It prints “Fizz” for even numbers, “Buzz” for numbers divisible by 5, and the number itself if neither condition is met.
All results are printed on one line.
