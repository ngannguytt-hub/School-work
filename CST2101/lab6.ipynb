### **TASK 1**

```python
import numpy as np

array = np.random.randint(10,51, size=(5,5))
print(array)

mean = np.mean(array)
median = np.median(array)
std = np.std(array)

maximum = np.max(array)
minimum = np.min(array)
max_idx = np.unravel_index(np.argmax(array), array.shape) 
min_idx = np.unravel_index(np.argmin(array), array.shape)

print(f'\nMean: {mean}, \nMedian: {median}, \nStandard Deviation: {std}')
print(f'Max: {maximum} at index {max_idx}, \nMin: {minimum} at index {min_idx}')
```

### **TASK 2**

```python
import scipy.stats as stats
import scipy.integrate as integrate

#Create a dataset
dataset = np.random.normal(loc=50, scale=10, size=100)

#Compute
mean_val = np.mean(dataset)
median_val = np.median(dataset)
mode_val = stats.mode(dataset, keepdims=True)
variance_val = np.var(dataset)
skewness_val = stats.skew(dataset)
kurtosis_val = stats.kurtosis(dataset)

print(f"Mean value: {mean_val}\n"
f"Median value: {median_val}\n"
f"Mode value: {mode_val}\n"
f"Variance: {variance_val}\n"
f"Skewness value: {skewness_val}\n"
f"Kurtosis value: {kurtosis_val}")
```
