### Problem 1
Handle the exception thrown by the code below by using <code>try</code> and <code>except</code> blocks.

Write a Python program to handle a ZeroDivisionError exception when dividing a number by zero.

```python
def div(a,b):
    try:
        c = a/b
        print(c)
    except ZeroDivisionError:
        print("Error: can't divide by zero")
```

```python
div(10,5)
```

```python
div(5,0)
```

### Problem 2

Write a Python program that prompts the user to input an integer and raises a ValueError exception if the input is not a valid integer.

```python
def get_integer_input(prompt):
    try:
        # Attempt to get an integer input from the user and store it in the 'value' variable.
        n = int(input(prompt))
    
        # Return the integer value.
        return n
       
    except ValueError:
        # Handle the exception if the user's input is not a valid integer.
       print("It's not a valid integer!")
```

```python
# Call the get_integer_input function to get an integer input from the user with the provided prompt.
a = get_integer_input("Please enter a valid integer: ")


# Print the input value obtained from the function.
print(f"You entered: {a}")
```

### Problem 4

Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical.

exception TypeError:

Raised when an operation or function is applied to an object of inappropriate type. The associated value is a string giving details about the type mismatch.

```python
def get2numbers(prompt1, prompt2):
        try:
            no1 = float(input(prompt1))
            no2 = float(input(prompt2))
            return no1, no2    
        except ValueError:
            raise TypeError("Inputs must be numerical")

try:
    no1, no2 = get2numbers("Please enter 1st number: ","Please enter 2nd number: ")
    print(f"You entered two numbers: {no1} and {no2}")
except TypeError as e:
    print("Error: ",e)
```

### Problem 5
Write a function that asks for an integer and prints the square of it. Use a <code>while</code> loop with a <code>try</code>, <code>except</code>, <code>else</code> block to account for incorrect inputs.

```python
def square():
    while True:
        try:
            n = int(input("Please enter a valid integer: "))
        except ValueError:
            print("Incorrect input. Please enter an integer")
            continue
        else:
            print("Square =", n**2)
            break
            
```

```python
square()
```

### Problem 7

Write a Python program that executes an operation on a list and handles an IndexError exception if the index is out of range.

exception IndexError:

Raised when a sequence subscript is out of range. (Slice indices are silently truncated to fall in the allowed range; if an index is not an integer, TypeError is raised.)

```python
list = [1,2,3,4]

try:
    idx = int(input("Enter an index: "))
    val = list[idx]
    print(val)
except IndexError:
    print("An index is out of range")
except ValueError:
    print("Error")
```

### Problem 8

Write a Python program that executes division and handles an ArithmeticError exception if there is an arithmetic error.

exception ArithmeticError:

The base class for those built-in exceptions that are raised for various arithmetic errors: OverflowError, ZeroDivisionError, FloatingPointError.

```python
try:
    no1 = float(input("Please enter the 1st number: "))
    no2 = float(input("Please enter the 2nd number: "))
    div = no1/no2
    print(div)
except ArithmeticError as e:
    print("Error: ",e)
        
```
